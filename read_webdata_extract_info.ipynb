# Perth Real Estate Market from 2009 - 2018

The purpose of this research is to find patterns in housing patterns in Perth real estate market. After the end on Mining boom in Perth in 2014 house prices fallen on average by 10% across all suburbs. However, not all suburbs folllowed the same trend.
Some suburbs lost as much as 25% of house value in 2018 since 2014, and prices continue to fall. On the other side, other, more expensiver suburbs continued growth and were profitable for investors. This data analysis will summarize patterns from the data.
Data for this analysis was coppied as it could be seen below with BeautifulSoup.
Visualizations for this analysis were done in Tableau and are available to see from my profile as a story:

## Bringing it all together

```python
#All Libraries used:   
#read csv file suburb,postcode
import pandas as pd
import numpy as np
import sys
from pandas.io.json import json_normalize
#import library tht is used for web parcing
from bs4 import BeautifulSoup
#library to download web pages
from urllib.request import Request, urlopen
#library to read json format
import json
import requests
import csv
#####################################

#read table that has suburb,postcode,state that are needed for URL creations
import_suburbs = pd.read_csv('suburb_postcode.csv')
# quick check
df=pd.DataFrame(import_suburbs)

myfile = open('xyz2.txt', 'w')
for pos, row in df.iterrows():
    #make urls with suburb data
    url='https://www.realestate.com.au/neighbourhoods/'
    url += row.c1 + '-'
    url += str(row.c2) + '-'
    url += row.c3
    #download the page and safe the result
    response = requests.get(url)
    req=Request(url,headers={'User-Agent': 'Mozilla/5.0'})
    webpage=urlopen(req).read()
    soup2=BeautifulSoup(webpage, 'html.parser')
    element='{"'+row.c1 + '":' + str(soup2.find('div', class_="slide-section median-price-subsections trend")) + '}"'
    print(element)
    myfile.write(element)
   # with open('file2.txt', 'w') as f:
    #    print(element, file=f)

myfile.close()    
    
```

## More data to extract for separate visualizations (house market)
So, the key median prices I want to extract for each suburb:

1a. price in 2011 (2011 was a year of previous low) 
1b. and number of houses sold in 2011

2a. price in 2014 (at peak for most suburbs)
2b. number of houses sold in 2014 per suburb

3a. price in sept 2018 (current low, best available data)
3b. number of houses sold from oct 2017- sept 2018 per suburb

Calculated fields:
4a change in % between 2011 and 2018(from oct 2017-sept 2018)

5a biggest gain in $ from 2011-2014
5b rise in % from 2011-2014

6a. drop in price from 2014-2018 (september) in $
6b. drop in price from 2014-2018 (september) in %




```python
#first convert house sales table to row index and groups/suburb into the columns
df_mean_house_price_conv=df_sales_price_houses.set_index(['date','suburb']).\
numbers.unstack(['suburb'])
df_mean_house_price_conv.tail(3)
```

```python
#1a - extracting price in 2011
house_price_2011=df_mean_house_price_conv.loc['31-12-11']
house_price_2011.head(3)
```

```python
#2a - extracting price in 2014
house_price_2014=df_mean_house_price_conv.loc['31-12-14']
house_price_2014.head(3)
```

```python
#3a - extracting price in 2018(september)
house_price_2018=df_mean_house_price_conv.loc['30-09-18']
house_price_2018.head(3)
```

```python
#first convert number of house sales table to row index and groups/suburb into the columns
#I have duplicates in this table, so just remove them
df_sales_number_houses_d = df_sales_number_houses.drop_duplicates(['date','category','suburb'])
#df_sales_number_houses_d.head(3)
#and unstack
df_mean_house_number_conv=df_sales_number_houses_d.set_index(['date','suburb']).\
numbers.unstack(['suburb'])
df_mean_house_number_conv.head(3)
```

```python
#1b. and number of houses sold in 2011
house_number_2011=df_mean_house_number_conv.loc['31-12-11']
house_number_2011.head(3)
```

```python
#2b - number of houses sold in 2014
house_number_2014=df_mean_house_number_conv.loc['31-12-14']
house_number_2014.head(3)
```

```python
#2b - number of houses sold from oct 2017- sept 2018 per suburb
house_number_2018=df_mean_house_number_conv.loc['30-09-18']
house_number_2018.head(3)
```

```python
#merging all data above in one dataframe
concat_all=pd.concat([house_price_2011, house_price_2014, house_price_2018,house_number_2011,\
                     house_number_2014, house_number_2018], axis=1)
```

```python
concat_all.head(3)
```

```python
#rename columns 
concat_all.columns=['house_price_2011','house_price_2014','house_price_2018',\
                   'house_number_2011','house_number_2014','house_number_2018']
concat_all.head(3)
```

```python
#Calculated fields:
#4a change in % between 2011 and 2018(from oct 2017-sept 2018)
concat_all['change_2011_2018per']=((concat_all['house_price_2011']-concat_all['house_price_2018'])\
                                   /concat_all['house_price_2011'])*100

#5b rise in % from 2011-2014
concat_all['change_2011_2014per']=((concat_all['house_price_2011']-concat_all['house_price_2014'])\
                                   /concat_all['house_price_2011'])*100

#5b сhange in % from 2014-2018
concat_all['change_2014_2018per']=((concat_all['house_price_2014']-concat_all['house_price_2018'])\
                                   /concat_all['house_price_2014'])*100

#5a biggest gain in $ from 2011-2014
concat_all['change_2011_2014dol']=concat_all['house_price_2011']-concat_all['house_price_2014']

#6a. drop in price from 2014-2018 (september) in $
concat_all['change_2014_2018dol']=concat_all['house_price_2014']-concat_all['house_price_2018']

concat_all.head(3)
```

```python
#write to csv file this table for tableau
concat_all.to_csv('realestate_change_2011_2018.csv', sep=',')
```
